```python
# This is the uploader widget

def _upload():

    _upload_widget = fileupload.FileUploadWidget()

    def _cb(change):
        global file_contents
        decoded = io.StringIO(change['owner'].data.decode('utf-8'))
        filename = change['owner'].filename
        print('Uploaded `{}` ({:.2f} kB)'.format(
            filename, len(decoded.read()) / 2 **10))
        file_contents = decoded.getvalue()

    _upload_widget.observe(_cb, names='data')
    display(_upload_widget)

_upload()
```

```python
def calculate_frequencies(file_contents):
    # Here is a list of punctuations and uninteresting words you can use to process your text
    punctuations = '''!()-[]{};:’‘”'"\,<>./?@#$%^&*_~'''
    uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just", \
    "about","into","in","on","went","above","then","put","go","so","after","not","for","up","jm","us","si","ve"]
    
    # LEARNER CODE START HERE
    #split txt file into individual words and characters
    text_list = file_contents.split()
    new_list = []
    interesting_words = []
    frequency={}
    
    #remove all punctuations from list created
    for item in text_list:
        newItem = item
        if item.isalpha():
            new_list.append(item.lower())
        else:
            for index in range(len(punctuations)):
                for eachItem in item: 
                    if punctuations[index] == eachItem:
                        newItem = newItem.replace(eachItem,"")    
            new_list.append(newItem.lower())
    
    
    
    #remove uninteresting words
    for item in new_list:
        if item not in uninteresting_words and len(item) > 2:
            interesting_words.append(item)
    
    #count the frequency of words appearing in interesting words
    for word in interesting_words:
        if word not in frequency:
            frequency.update({word : 1})
        else:
            frequency[word] =frequency[word]+1
        
    
    #wordcloud
    cloud = wordcloud.WordCloud()
    cloud.generate_from_frequencies(frequency)
    return cloud.to_array()
```

```python
# Display your wordcloud image
with open("wordCloud.txt", "r") as f:
    file_contents = f.read()
    
myimage = calculate_frequencies(file_contents)
plt.imshow(myimage, interpolation = 'nearest')
plt.axis('off')
plt.savefig('word_cloud.jpeg')
plt.show()
```
